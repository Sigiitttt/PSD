# **1. Data Understanding - iris**

Tahap pemahaman data bertujuan untuk mengenal karakteristik dasar dari dataset yang digunakan. Ini melibatkan pemahaman tentang sumber data, arti dari setiap kolom (fitur), dan statistik dasar untuk mendapatkan gambaran umum sebelum melangkah ke analisis yang lebih dalam.

#### **a. Sumber dan Pemuatan Data**

* **Sumber Data**: Dataset yang digunakan adalah **Iris Flower Dataset**, sebuah dataset klasik dalam dunia statistika dan *machine learning* yang diperkenalkan oleh Ronald Fisher pada tahun 1936. Dataset ini berisi data pengukuran dari tiga spesies bunga Iris.
* **Proses Pemuatan**: Untuk proyek ini, data pertama kali dimuat ke dalam database **MySQL** dan **PostgreSQL**. Selanjutnya, data ditarik ke **Power BI** menggunakan konektor database untuk dianalisis dan divisualisasikan.

#### **b. Deskripsi Atribut/Kolom**

Dataset ini terdiri dari 150 sampel data dengan 5 atribut utama.

* `SepalLengthCm` (Tipe: Desimal): Panjang kelopak luar bunga dalam satuan sentimeter.
* `SepalWidthCm` (Tipe: Desimal): Lebar kelopak luar bunga dalam satuan sentimeter.
* `PetalLengthCm` (Tipe: Desimal): Panjang kelopak dalam bunga dalam satuan sentimeter.
* `PetalWidthCm` (Tipe: Desimal): Lebar kelopak dalam bunga dalam satuan sentimeter.
* `Species` (Tipe: Teks): Nama spesies bunga, yang terdiri dari tiga kelas unik:
    1.  `Iris-setosa`
    2.  `Iris-versicolor`
    3.  `Iris-virginica`

#### **c. Analisis Data Eksplorasi **

Analisis awal dilakukan menggunakan Power BI untuk mendapatkan wawasan statistik dan visual.

**i. Statistik Deskriptif**

Tabel berikut merangkum statistik utama (nilai minimum, maksimum, dan rata-rata) untuk setiap fitur pengukuran.

| Fitur | Minimum (cm) | Maksimum (cm) | Rata-rata (cm) |
| :--- | :--- | :--- | :--- |
| **Sepal Length** | 4.30 | 7.90 | 5.84 |
| **Sepal Width** | 2.00 | 4.40 | 3.05 |
| **Petal Length** | 1.00 | 6.90 | 3.76 |
| **Petal Width** | 0.10 | 2.50 | 1.20 |

**ii. Distribusi Kelas**

Visualisasi distribusi kelas menunjukkan bahwa dataset ini **seimbang (*balanced*)**. Setiap spesies memiliki jumlah sampel yang sama persis.



* **Iris-setosa**: 50 sampel
* **Iris-versicolor**: 50 sampel
* **Iris-virginica**: 50 sampel

Distribusi yang seimbang ini sangat ideal untuk tugas klasifikasi, karena model *machine learning* tidak akan cenderung bias pada salah satu kelas.

#### **d. Temuan Utama dan Wawasan Awal**

1.  **Rentang Ukuran**: Fitur kelopak dalam (`Petal Length` & `Petal Width`) memiliki rentang nilai yang lebih lebar dan nilai minimum yang jauh lebih kecil dibandingkan kelopak luar (`Sepal Length` & `Sepal Width`). Hal ini mengindikasikan bahwa ukuran kelopak dalam mungkin menjadi pembeda yang kuat antar spesies.
2.  **Kualitas Data**: Dataset ini bersih dan lengkap (tidak ada nilai yang hilang) serta memiliki distribusi kelas yang seimbang, sehingga sangat siap untuk tahap pemodelan lebih lanjut.
3.  **Visualisasi Efektif**: Penggunaan Power BI memungkinkan eksplorasi statistik kunci secara cepat dan efisien melalui visualisasi interaktif seperti *Card* dan *Bar Chart*.
