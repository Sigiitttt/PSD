# **Data Understanding: Iris Dataset**

#### **Pendahuluan**

Dataset Iris adalah salah satu dataset klasik dalam bidang data science dan machine learning. Dataset ini pertama kali diperkenalkan oleh ahli statistik Ronald Fisher pada tahun 1936 sebagai contoh untuk analisis diskriminan linier. Dataset ini berisi pengukuran dari 150 bunga iris dari tiga spesies: *Iris-setosa*, *Iris-versicolor*, dan *Iris-virginica*. Setiap sampel memiliki empat fitur pengukuran (dalam cm): panjang sepal (sepal_length), lebar sepal (sepal_width), panjang petal (petal_length), dan lebar petal (petal_width), serta label spesies.

Dataset ini berisi data yang mencakup header dan 150 baris data (50 per spesies). Tidak ada nilai hilang (missing values), dan dataset ini seimbang (balanced).

Sumber dataset: Kaggle (https://www.kaggle.com/datasets/data855/heart-disease?resource=download)

#### **Deskripsi Data**
- **Jumlah Baris**: 150 (50 untuk setiap spesies).
- **Jumlah Kolom**: 5.
- **Kolom dan Tipe Data**:
  - `sepal_length`: Numerik (float), panjang sepal dalam cm.
  - `sepal_width`: Numerik (float), lebar sepal dalam cm.
  - `petal_length`: Numerik (float), panjang petal dalam cm.
  - `petal_width`: Numerik (float), lebar petal dalam cm.
  - `species`: Kategorikal (string), nama spesies bunga (Iris-setosa, Iris-versicolor, Iris-virginica).
- **Missing Values**: Tidak ada (0 untuk semua kolom).
- **Distribusi Kelas**: Seimbang, dengan 50 sampel per spesies.
- **Duplikat**: Tidak ada duplikat signifikan berdasarkan analisis awal.

Dataset ini tidak memiliki outlier (nilai aneh) yang terlalu ekstrem, tapi ada variasi wajar antar jenis.

# **Analisis Data dan Identifikasi Outlier pada Dataset Iris**

### **1. Pemahaman Data (Data Understanding)**

Dataset yang digunakan adalah dataset klasik **Iris**, berisi 150 sampel dengan 4 fitur numerik:

* **Sepal Length** (panjang kelopak)
* **Sepal Width** (lebar kelopak)
* **Petal Length** (panjang mahkota)
* **Petal Width** (lebar mahkota)

Selain itu terdapat kolom `id` (identitas unik) dan `Class` (label spesies: *Setosa*, *Versicolor*, *Virginica*).

Dalam proses deteksi outlier:

* Kolom `id` diabaikan karena hanya identitas unik.
* Kolom `Class` tidak digunakan, sebab metode yang dipakai bersifat **unsupervised** (tanpa label).

Dataset akhir yang dianalisis berisi **150 baris  4 fitur numerik**.

## **2. Metode Deteksi Outlier**

Metode yang digunakan adalah **ABOD (Angle-Based Outlier Detection)**.

* ABOD menilai titik data berdasarkan **distribusi sudut** yang dibentuk dengan titik-titik lain.
* Outlier ditandai ketika distribusi sudutnya jauh berbeda dibanding titik normal.
* Kelebihan ABOD: efektif mendeteksi outlier dalam data berdimensi rendah hingga sedang.

Dengan **PyCaret 3.2/3.3**, langkah-langkahnya:

1. Setup environment anomaly detection.
2. Membuat model ABOD.
3. Menambahkan kolom `Anomaly` ke dataset:

   * `0` = data normal
   * `1` = data outlier

---

## **3. Hasil Analisis Outlier**

Setelah menjalankan ABOD, diperoleh distribusi anomali berikut:

* **Normal (0)** : 142 data

 **Catatan:**
Hasil deteksi menunjukkan sebagian kecil data Iris memiliki pola berbeda. Outlier bisa jadi **anomali alami** dari variasi spesies, atau data yang perlu ditinjau ulang. Tahap ini penting agar analisis dan model prediktif berikutnya tidak bias oleh data ekstrem.


#### **cuplikan code**

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pycaret.anomaly import *

# 1. Load Data
data = pd.read_csv("data_iris.csv")

print("=== Informasi Dataset ===")
print(data.info())
print("\n=== 5 Data Teratas ===")
print(data.head())

# 2. Setup PyCaret (hanya fitur numerik)
features = data.drop(columns=["id", "Class"])
setup(data=features, session_id=123, verbose=False)

abod = create_model("abod")
results_abod = assign_model(abod)

# Tambahkan hasil ke dataframe asli
data_abod = data.copy()
data_abod["Anomaly"] = results_abod["Anomaly"]

print("\n=== Distribusi Outlier (ABOD) ===")
print(data_abod["Anomaly"].value_counts())  # 0 = normal, 1 = outlier

# Hitung jumlah normal & outlier
n_outliers = (data_abod["Anomaly"] == 1).sum()
n_normal = (data_abod["Anomaly"] == 0).sum()

# Plot normal (hijau bulat)
sns.scatterplot(
    x=data_abod.loc[data_abod["Anomaly"]==0, "sepal length"],
    y=data_abod.loc[data_abod["Anomaly"]==0, "sepal width"],
    color="blue", alpha=0.6, s=50,
    label=f"Normal ({n_normal})"
)

# Plot outlier (orange X besar)
sns.scatterplot(
    x=data_abod.loc[data_abod["Anomaly"]==1, "sepal length"],
    y=data_abod.loc[data_abod["Anomaly"]==1, "sepal width"],
    color="red", marker="X", s=120, alpha=0.9,
    label=f"Outlier ({n_outliers})"
)

plt.title("Identifikasi Outlier dengan ABOD")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend(title="Kategori")
plt.grid(alpha=0.3)
plt.savefig("outlier_abod.png")
plt.show()
````
