# **Business Understanding – Coffee Shop Sales**

### 1. **Latar Belakang Masalah**

Maven Roasters adalah sebuah coffee shop fiktif dengan tiga lokasi di New York City. Seperti bisnis retail lainnya, memahami perilaku penjualan sangat penting untuk:

* Meningkatkan pendapatan.
* Mengoptimalkan persediaan produk.
* Mengatur strategi promosi.
  Data transaksi ini mencakup tanggal, waktu, lokasi toko, serta detail produk yang dijual.

---

### 2. **Tujuan Bisnis**

* **Memahami tren penjualan** dari waktu ke waktu untuk mengetahui pola pertumbuhan atau penurunan pendapatan.
* **Menganalisis perilaku pelanggan**: kapan (hari/waktu) toko paling ramai, dan produk apa yang paling diminati.
* **Mengidentifikasi produk andalan** (produk dengan volume tinggi atau kontribusi pendapatan terbesar).
* **Mendukung pengambilan keputusan bisnis** terkait stok, harga, promosi, dan strategi penjualan.

---

### 3. **Pertanyaan Bisnis**

1. Bagaimana tren penjualan Maven Roasters dari waktu ke waktu?
2. Hari apa dalam seminggu yang paling ramai, dan apakah ada pola tertentu (weekday vs weekend)?
3. Produk apa yang paling laku dan mana yang jarang terjual?
4. Produk atau kategori mana yang menyumbang pendapatan terbesar?
5. Apakah ada perbedaan pola penjualan antar lokasi coffee shop?

---

### 4. **Ruang Lingkup**

* **Data Input:**

  * Transaksi: ID, tanggal, waktu, jumlah, lokasi toko.
  * Produk: kategori, jenis, detail, harga.
* **Output Analisis:**

  * Insight tentang pola penjualan, perilaku pelanggan, dan kontribusi produk terhadap pendapatan.
* **Batasan:**

  * Data hanya mencakup 3 lokasi coffee shop di NYC.
  * Tidak ada data tentang profil pelanggan individu (usia, gender, preferensi).

---

### 5. **Nilai Bisnis (Business Value)**

* **Manajemen toko** → bisa mengatur jadwal karyawan sesuai jam sibuk.
* **Tim inventori** → tahu produk mana yang perlu stok lebih banyak atau lebih sedikit.
* **Marketing** → dapat menyusun strategi promosi sesuai produk/harian tertentu.
* **Owner** → memahami tren jangka panjang untuk ekspansi atau strategi harga.

---

⚡ Jadi, dari dataset ini tujuan akhirnya adalah memberikan **insight berbasis data** agar bisnis coffee shop bisa **lebih efisien, responsif, dan profitabel**.
