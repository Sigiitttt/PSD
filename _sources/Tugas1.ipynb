## 1. 🎯 Memahami Bisnis (Business Understanding)  

### a. Definisi
Memahami bisnis adalah tahap awal dalam **CRISP-DM (Cross-Industry Standard Process for Data Mining)**.  
Tujuannya: mengetahui **apa masalah bisnis yang ingin diselesaikan** dan **apa tujuan yang ingin dicapai** dengan data science.

### b. Langkah-langkah Business Understanding
1. **Identifikasi Masalah Bisnis**  
   - Apa tantangan atau masalah yang dihadapi organisasi?  
   - Contoh: penurunan penjualan, churn pelanggan tinggi, biaya operasional meningkat.

2. **Tentukan Tujuan Proyek**  
   - Rumuskan tujuan dalam bahasa bisnis.  
   - Contoh: meningkatkan penjualan 20%, menurunkan churn pelanggan 10%, optimasi biaya 15%.  

3. **Formulasi Masalah Data**  
   - Terjemahkan masalah bisnis ke bentuk masalah data.  
   - Contoh: "Bagaimana kita memprediksi pelanggan yang akan churn?" → menjadi **klasifikasi**.  

4. **Tentukan Kriteria Keberhasilan**  
   - Gunakan **ukuran bisnis** dan **ukuran teknis**.  
   - Contoh: AUC > 80% + churn turun 10% dalam 6 bulan.

---

## 2. 🔍 Data Understanding  

### a. Definisi  
Data understanding adalah proses **mengumpulkan, mengenali, dan memahami data** yang akan digunakan untuk menjawab masalah bisnis.

### b. Langkah-langkah Data Understanding
1. **Kumpulkan Data Awal**  
   - Ambil dataset dari database, file CSV, API, dll.  

2. **Deskripsi Data**  
   - Apa saja fitur/kolom yang tersedia?  
   - Tipe data: numerik, kategorikal, teks, tanggal, dll.  

3. **Eksplorasi Awal**  
   - Lihat ukuran data (baris, kolom).  
   - Lihat statistik deskriptif (mean, median, nilai unik).  

4. **Identifikasi Masalah Data**  
   - Missing values, outliers, duplikasi, data tidak konsisten.
