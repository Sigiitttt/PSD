# **Data Understanding – Coffee Shop Sales**

### 1. **Sumber Data**

* Dataset: *Coffee Shop Sales (Maven Roasters [https://www.kaggle.com/datasets/ahmedabbas757/coffee-sales])* dari Kaggle.
* Data ini merepresentasikan transaksi penjualan di tiga lokasi coffee shop di New York City.
* Tujuan: untuk analisis perilaku penjualan dan praktik analisis bisnis.

---

### 2. **Deskripsi Atribut**

| No | Atribut            | Tipe Data     | Deskripsi                                                                  |
| -- | ------------------ | ------------- | -------------------------------------------------------------------------- |
| 1  | `transaction_id`   | Integer       | ID unik tiap transaksi                                                     |
| 2  | `transaction_date` | Date          | Tanggal transaksi (format MM/DD/YY)                                        |
| 3  | `transaction_time` | Time          | Waktu transaksi (format HH\:MM\:SS)                                        |
| 4  | `transaction_qty`  | Integer       | Jumlah item yang dibeli dalam satu transaksi                               |
| 5  | `store_id`         | Integer       | ID unik lokasi coffee shop                                                 |
| 6  | `store_location`   | Kategori      | Nama lokasi coffee shop (contoh: Lower Manhattan, Astoria, Hell’s Kitchen) |
| 7  | `product_id`       | Integer       | ID unik produk yang dijual                                                 |
| 8  | `unit_price`       | Numerik       | Harga per unit produk (USD)                                                |
| 9  | `product_category` | Kategori      | Kategori produk (contoh: Coffee, Tea, Bakery, Merchandise)                 |
| 10 | `product_type`     | Kategori      | Jenis produk dalam kategori (contoh: Espresso, Latte, Muffin, Tumbler)     |
| 11 | `product_detail`   | Teks/Kategori | Detail spesifik produk (contoh: “Espresso - Small”, “Blueberry Muffin”)    |

---

### 3. **Kualitas Data**

* **Completeness (Kelengkapan):** dataset relatif lengkap, tidak ada atribut kosong yang penting.
* **Consistency (Konsistensi):** format tanggal & waktu konsisten, meski mungkin perlu diubah jadi tipe datetime untuk analisis.

* **Validity (Validitas):**

  * `transaction_qty` seharusnya bernilai positif.
  * `unit_price` juga tidak boleh bernilai negatif.
* **Granularity (Kedetailan):** data tersedia sampai ke level **produk per transaksi**, sehingga bisa dianalisis dari sisi penjualan harian, mingguan, lokasi, hingga produk.

---

### 4. **Eksplorasi Awal (EDA Sederhana)**

Beberapa hal yang bisa dieksplor dari dataset ini:

1. **Tren Penjualan Harian/Mingguan**

   * Hitung total pendapatan per hari → lihat pola naik/turun.
2. **Hari Tersibuk**

   * Kelompokkan transaksi berdasarkan hari dalam seminggu.
   * Bandingkan weekdays vs weekends.
3. **Produk Paling Laris & Paling Sepi**

   * Hitung total penjualan berdasarkan `product_type` dan `product_category`.
4. **Pendapatan per Lokasi**

   * Bandingkan performa penjualan tiap `store_location`.

---

### 5. **Insight Awal (Hipotesis)**

* Penjualan kemungkinan lebih tinggi di **weekday pagi** (jam orang kerja beli kopi).
* Produk **kopi (Espresso, Latte, Cappuccino)** mungkin jadi kontributor pendapatan terbesar.
* Lokasi di daerah **kantor (misalnya Lower Manhattan)** bisa lebih ramai dibanding area lain.
* Produk **merchandise** (contoh: tumbler, mug) mungkin jarang terjual tapi harganya tinggi → kontribusi ke revenue lumayan.
