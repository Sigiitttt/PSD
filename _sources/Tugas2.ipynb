## **Data Understanding**




Setelah kita tahu **apa masalah bisnis** yang mau diselesaikan di tahap *Business Understanding*, sekarang saatnya kita berkenalan lebih dekat dengan "bahan baku" utama kita: **Data**.

Selamat datang di tahap kedua: **Data Understanding** (Memahami Data).

Jika di tahap sebelumnya kita menjadi "dokter bisnis", sekarang kita bertransformasi menjadi "detektif". Tugas kita adalah menginvestigasi data yang kita miliki, mencari petunjuk, memahami karakternya, dan menemukan keanehan-keanehan yang mungkin tersembunyi.

> **Analogi Sederhana: Memasak dengan Bahan Baru**
>
> Bayangkan Anda seorang koki yang diberi sekotak bahan makanan misterius untuk dimasak. Apa yang akan Anda lakukan?
>
> Tentu Anda tidak akan langsung mencampurnya ke dalam panci. Anda akan membukanya satu per satu, mencium aromanya, mungkin mencicipi sedikit, melihat teksturnya, dan memisahkan mana bahan yang segar dan mana yang sudah kurang bagus.
>
> Tahap **Data Understanding** adalah persis seperti itu. Kita "mencicipi" dan "merasakan" data kita sebelum mengolahnya lebih lanjut.

Tujuan dari tahap ini adalah untuk membangun pemahaman yang kuat tentang data, yang akan menjadi dasar untuk semua langkah teknis selanjutnya.

---

**Langkah-langkah Praktis dalam Memahami Data (Lanjutan Studi Kasus: Kedai Kopi "Senja")**

Kita akan melanjutkan investigasi kita pada data dari **Kedai Kopi Senja**. Di tahap sebelumnya, kita tahu bahwa kita punya data transaksi dan data pelanggan.

**Langkah 1: Mengumpulkan Data Awal (Initial Data Collection)**

Ini adalah langkah paling awal: mengumpulkan semua data yang relevan dari berbagai sumber ke satu tempat.

* **Tindakan:** Kita meminta akses ke database mesin kasir dan sistem program loyalitas Kedai Kopi Senja. Kita *meng-export* data tersebut, misalnya dalam format file CSV atau Excel.
* **Hasil:** Kita sekarang punya dua file utama: `transaksi.csv` dan `pelanggan.csv`.

**Langkah 2: Mendeskripsikan Data (Describe Data)**

Setelah data terkumpul, kita perlu mendokumentasikan "kartu identitas" dari setiap set data.

* **Tindakan:** Kita periksa kedua file tersebut.
    * **Untuk `transaksi.csv`:** Ada berapa baris (jumlah transaksi)? Ada berapa kolom? Apa saja nama kolomnya (`id_transaksi`, `tanggal`, `item`, `id_pelanggan`)? Apa tipe data di setiap kolom (teks, angka, tanggal)?
    * **Untuk `pelanggan.csv`:** Sama, kita periksa jumlah baris (jumlah pelanggan), nama kolom, dan tipe datanya.
* **Hasil:** Kita punya gambaran umum tentang volume dan format data. Misalnya, "Terdapat 15.000 baris transaksi dalam 6 bulan terakhir dan 850 pelanggan terdaftar."

**Langkah 3: Menjelajahi Data (Explore Data)**

Inilah bagian paling seru dari pekerjaan detektif! Kita mulai "menginterogasi" data dengan statistik dan visualisasi sederhana untuk mencari pola atau wawasan awal.

* **Tindakan:**
    1.  **Statistik Deskriptif:** Kita lihat statistik dasar dari data numerik. Misalnya, berapa rata-rata nilai transaksi? Berapa transaksi terbesar dan terkecil?
    2.  **Melihat Distribusi:** Produk apa yang paling laris? Hari apa yang paling ramai (Senin, Minggu)? Jam berapa puncak keramaian kedai kopi? Kita bisa membuat grafik batang sederhana untuk melihatnya.
    3.  **Mencari Hubungan Awal:** Apakah ada hubungan antara hari dan jenis produk yang dibeli? (misalnya, di akhir pekan lebih banyak yang beli minuman non-kopi?).
* **Hasil (Contoh Temuan):**
    * "Wow, ternyata **Kopi Susu Gula Aren** adalah produk terlaris, terjual 3x lebih banyak dari Americano."
    * "Puncak keramaian terjadi pada jam 8-10 pagi dan jam 3-5 sore."
    * "Transaksi di hari Sabtu dan Minggu rata-rata nilainya lebih tinggi daripada hari kerja."

**Langkah 4: Memverifikasi Kualitas Data (Verify Data Quality)**

Tidak ada data yang sempurna. Di tahap ini, kita mencari "cacat" atau masalah dalam data kita.

* **Tindakan:**
    1.  **Mencari Nilai yang Hilang (*Missing Values*):** Apakah ada transaksi yang tidak tercatat `id_pelanggan`-nya? Kenapa?
    2.  **Mencari Kesalahan atau Keanehan (*Anomalies/Errors*):** Apakah ada nama produk yang salah ketik (misal: "Croissant", "Croisant", "Crossant")? Apakah ada nilai transaksi yang minus atau nol?
    3.  **Memeriksa Konsistensi:** Apakah format tanggal sudah seragam di semua baris?
* **Hasil (Contoh Temuan):**
    * "Ditemukan 20% transaksi tidak memiliki `id_pelanggan`. Kemungkinan ini adalah transaksi dari pelanggan non-member."
    * "Nama produk 'Americano' terkadang ditulis sebagai 'Americano Black'. Ini harus diseragamkan."
    * "Ada satu transaksi dengan nilai Rp 5.000.000, ini tidak wajar. Perlu diperiksa apakah ini salah input."

---

Kesimpulan: Kenapa Tahap Ini Penting?

Melewatkan tahap **Data Understanding** adalah resep menuju bencana. Tanpa memahami data kita, kita bisa saja:
* Membuat model dari data yang salah atau "kotor", yang akan menghasilkan kesimpulan yang salah (*garbage in, garbage out*).
* Kehilangan wawasan penting yang sebenarnya sudah terlihat dari eksplorasi data sederhana.
* Menghabiskan waktu berhari-hari untuk memperbaiki masalah di tahap selanjutnya, padahal masalah itu bisa dideteksi lebih awal di sini.

Dengan menyelesaikan tahap ini, kita tidak hanya memahami data, tetapi juga mulai membangun intuisi tentang cerita apa yang bisa diungkapkan oleh data tersebut. Kita siap untuk tahap selanjutnya: **menyiapkan data untuk dianalisis**.
